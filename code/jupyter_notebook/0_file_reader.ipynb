```python
from shapely.geometry import Polygon
from hierarchical_grouping.file_reader import BinaryPolygonFileReader
```

# 1. Reader 
we read data using BinaryPolygonFileReader from the module.

```python
file_path = "../data/data_zentrum"
reader = BinaryPolygonFileReader(file_path)
```

# 2. Method inside the reader
Reader contains one method which is read_next_set. It read the data inside until it runs out of data. When it run out of data it will run None. That mean, in order to get all data, you need to run it iteratively.

```python
set_id, polys1, polys2 = reader.read_next_set()

print("set id: ",set_id)

# Output the results
print("First Set of Polygons:")
for poly in polys1:
    print(poly)

print("\nSecond Set of Polygons:")
for poly in polys2:
    print(poly)
```
